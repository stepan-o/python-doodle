```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_xy(x, y,
            title="",
            create=True, show=True, ax=None,
            v_line=None, h_line=None,
            sup_line_style ='--', sup_line_width = 0.5, sup_line_color ='gray',
            x_lim = None):
    """
    plot funcs
    """

    def plot_lines(type, line):
        """
        plots and annotates horizontal and vertical lines
        """
        if type == 'h':
            ax.axhline(line, color=sup_line_color,
                       linestyle=sup_line_style, linewidth=sup_line_width)
            ax.text(0, line, line)
        elif type == 'v':
            ax.axvline(line, color=sup_line_color,
                       linestyle=sup_line_style, linewidth=sup_line_width)
            ax.text(line, 0, line)
        else:
            print("func 'plot_lines': Input parameter 'type' must be either 'h' or 'v'.")

    if create == True:
        f, ax = plt.subplots(1, figsize=(6, 6))
        f.suptitle(title)
        # plot axes
        ax.axvline(0, color='black', linestyle='-', linewidth=1)
        ax.axhline(0, color='black', linestyle='-', linewidth=1)
        ax.grid(False)


    # plot the function
    ax.plot(x, y)
    #ax.text(1, 0.1, "$P(A_n)$")
    ax.set_xlabel('x')
    #ax.set_xticks(np.arange(max_trials + 1))
    ax.set_ylabel('y', rotation=0)

    if type(v_line) is list:
        for line in v_line:
            plot_lines('v', line)
    elif type(v_line) is float or type(v_line) is int:
        plot_lines('v', v_line)

    if type(h_line) is list:
        for line in h_line:
            plot_lines('h', line)
    elif type(h_line) is float or type(h_line) is int:
        plot_lines('h', v_line)

    # zoom axis
    if x_lim: ax.set_xlim(x_lim[0], x_lim[1])


    if show == True:
        plt.show()
```

```python
x = np.linspace(-10, 10, 1000)
y = np.sin(x)
title = ("sin(x)")
plot_xy(x, y, title)
```

```python
x = np.linspace(-10, 10, 1000)
y = np.sin(x) / x
title = ("sinx / x")
plot_xy(x, y, title)
```

```python
x = np.linspace(-10, 10, 1000)
y = np.sin(x) * x
title = ("$sinx \cdot{ x }$")
plot_xy(x, y, title)
```

```python
x = np.linspace(-10, 10, 1000)
y = np.sin(1 / x)
title = ("$sin (1/x) $")
plot_xy(x, y, title)
```

```python
x = np.linspace(-1, 1, 10000)
y = np.sin(1 / x)
title = ("$sin (1 / x) $")
plot_xy(x, y, title)
```

```python
x = np.linspace(-1, 1, 10000)
y = np.sin(1 / x) * x
title = ("$sin (1 / x) \cdot{x} $")
plot_xy(x, y, title)
```

```python
x = np.linspace(-0.1, 0.1, 10000)
y = np.sin(1 / x) * x
title = ("$sin (1 / x) \cdot{x} $")
plot_xy(x, y, title, show=False)
ax = plt.gca()
y = x
plot_xy(x, y, ax=ax, create=False, show=False)
y = -x
plot_xy(x, y, ax=ax, create=False, show=True)
```

```python
x = np.linspace(-0.001, 0.001, 1000)
y = np.sin(1 / x) * x
title = ("$sin (1 / x) \cdot{x} $ -- limit at x -> 0 is 0")
plot_xy(x, y, title, show=False)
ax = plt.gca()
y = x
plot_xy(x, y, ax=ax, create=False, show=False)
y = -x
plot_xy(x, y, ax=ax, create=False, show=True)
```

```python
x = np.linspace(-10, 10, 10000)
y = x ** 2 + 2 * x - 8
title = ("$x ^ 2 + 2  x - 8$")
plot_xy(x, y, title)
```

```python
x = np.linspace(-10, 10, 10000)
y = x - 2
title = ("$x - 2$")
plot_xy(x, y, title)
```

```python
x = np.linspace(-10, 10, 10000)
y = (x ** 2 + 2 * x - 8) / (x - 2)
title = ("$(x ** 2 + 2 * x - 8) / (x - 2)$")
plot_xy(x, y, title)
```

```python
x = np.linspace(4, 8, 10000)
y = (x ** 2 + 2 * x - 8) / (x - 5)
title = ("$(x ** 2 + 2 * x - 8) / (x - 5)$")
zoom_x = [4.99, 5.01]
plot_xy(x, y, title, x_lim=zoom_x)
```
