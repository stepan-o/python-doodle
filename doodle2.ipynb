```python
import numpy as np
import matplotlib.pyplot as plt
```

## Linear search vs binary search

```python
num_samples = np.array([1, 5, 10, 100, 300, 500, 1000, 10000, 100000, 200000, 300000, 
                        400000, 500000, 600000, 700000, 800000, 900000, 1000000])
num_questions = np.floor(np.log2(num_samples)) + 1
plt.plot(num_samples, num_questions, label='binary search')
plt.plot(num_samples, num_samples, label='linear search')
plt.xscale('log')
plt.title("Binary search vs linear search")
plt.xlabel("Number of samples to search")
plt.ylabel("Number of comparisons to find target")
```

```python
num_samples = 100000000
res = num_samples / np.floor(np.log2(num_samples)) + 1
print("On {0:,} samples, ratio of number of comparisons made by worst-case linear search vs binary: {1:,.2f} "
      .format(num_samples, res))
```
